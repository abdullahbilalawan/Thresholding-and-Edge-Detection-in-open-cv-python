```python
import cv2  as c
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = c.imread('obama.jpg')
```

```python
plt.imshow(image)
```

```python
#sobel filter
```

```python
# Extract Sobel Edges
sobel_x = c.Sobel(image, c.CV_64F, 0, 1, ksize=5)
sobel_y = c.Sobel(image, c.CV_64F, 1, 0, ksize=5)
```

```python
plt.imshow(sobel_x)
```

```python
plt.imshow(sobel_y)
```

```python
# Canny Edge Detection uses gradient values as thresholds
# The first threshold gradient
canny = c.Canny(image, 50, 120)
```

```python
plt.imshow(canny)
```
